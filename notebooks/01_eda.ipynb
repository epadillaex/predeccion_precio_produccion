# Exploración y análisis inicial

En este notebook se realizará la exploración y análisis inicial de un conjunto de datos sintético diseñado para el desarrollo de un modelo predictivo del coste de producción de memorias RAM.

El dataset utilizado ha sido generado de forma sintética mediante inteligencia artificial con el objetivo de simular características técnicas y comerciales habituales en módulos de memoria RAM. Entre las variables incluidas se encuentran atributos como la marca, tipo de memoria (DDR4 o DDR5), capacidad, frecuencia, latencia, número de módulos y otras características técnicas que influyen en el coste de producción del producto.

El propósito de este análisis exploratorio es comprender la estructura y comportamiento de los datos antes de construir el modelo predictivo. Para ello se realizarán diferentes tareas de análisis y validación del dataset.

## 1. Importación de bibliotecas y carga de datos

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
```

```python
ruta = os.path.join(os.path.dirname(os.getcwd()), 'data', 'raw', 'ram_price_dataset_2000.csv')
df = pd.read_csv(ruta)
display(df.head())
```

## 2. Revisión del dataset

```python
print("Nº de filas: ", df.shape[0])
print("Nº de columnas: ", df.shape[1])
```

```python
print("\nTipos de datos:")
print(df.dtypes)
```

El dataset contiene 2000 muestras, cada muestra tiene nueve características técnicas y la variable objetivo `Precio_USD`.

## 3. Análisis exploratorio de los datos

```python
df.describe()
```

```python
# Distribución del valor total
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(df["Capacidad_GB"], bins=30, ax=ax)
ax.set_title("Distribución de la 'Capacidad (GB)'")
ax.set_xlabel("Capacidad (GB)")
plt.tight_layout()
plt.show()
```

```python
# Distribución del valor total
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(df["Velocidad_MHz"], bins=30, ax=ax)
ax.set_title("Distribución de la 'Velocidad (MHz)'")
ax.set_xlabel("Velocidad (MHz)")
plt.tight_layout()
plt.show()
```

```python
# Distribución del valor total
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(df["Latencia_CAS"], bins=30, ax=ax)
ax.set_title("Distribución de la 'Latencia (CAS)'")
ax.set_xlabel("Latencia (CAS)")
plt.tight_layout()
plt.show()
```
